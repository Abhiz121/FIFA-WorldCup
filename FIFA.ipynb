```python
mlt.figure(figsize=(10, 6))
sns.histplot(DATA['Year'], bins=30, kde=True)
mlt.title('Matches in Year')
mlt.show()
```

```python
mlt.figure(figsize=(10, 6))
mlt.plot(DATA['Year'], DATA['Winner'], marker='o')
mlt.title('Most Match Won By Country')
mlt.show()
```

```python
mlt.figure(figsize=(15, 6))
sns.histplot(DATA['Runners-Up'], bins=30, kde=True)
mlt.title('Runners-up')
mlt.show()
```

```python
mlt.figure(figsize=(15, 6))
sns.barplot(x='QualifiedTeams',y='Country',data=DATA)
mlt.title('Qualified Teams')
mlt.show()
```

```python
mlt.figure(figsize=(10, 6))
sns.barplot(x='GoalsScored',y='Country',data=DATA,hue="Country")
mlt.title('Goal Scored')
mlt.show()
```
